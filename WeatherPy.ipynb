```python
import pandas as pd
import numpy as np
import requests, json
import matplotlib.pyplot as plt
from citipy import citipy
import openweathermapy.core as owm
```

```python
from config import api_key
```

```python
#Generate Random Lat and Long data to fetch random cities from citipy
random_lat = np.random.randint(-90, 91, 2000)
#print(len(random_lat))
random_long = np.random.randint(-180, 181, 2000)
# print(random_long)
```

```python
random_cities = []
for gc in range(len(random_lat)):
    nearest_city = citipy.nearest_city(random_lat[gc], random_long[gc])
    if nearest_city not in random_cities:
        random_cities.append(nearest_city)

#print(random_cities)
# type(nearest_city)
# print(len(random_cities))
# type(random_cities)
# print(random_cities[0].city_name)
```

```python
#Initlialize lists to store the values returned by OWM API
temperature_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
lat_list = []
long_list = []
city_list = []
country_list = []
```

```python
#Fetch the weather data from OWM API
#set base url for OWM
base_url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid={api_key}&q="
# print (base_url)
for city_id in range(len(random_cities)):
    print(f"------Fetching weather info for city #{city_id}, {random_cities[city_id].city_name}------")
    complete_url = f"{base_url}{random_cities[city_id].city_name},{random_cities[city_id].country_code}"
    try:
        query = requests.get(complete_url)
        response = query.json()
        temperature_list.append(response['main']['temp_max'])
        humidity_list.append(response['main']['humidity'])
        cloudiness_list.append(response['clouds']['all'])
        wind_speed_list.append(response['wind']['speed'])
        lat_list.append(response['coord']['lat'])
        long_list.append(response['coord']['lon'])
        city_list.append(response['name'])
        country_list.append(response['sys']['country'])
    except KeyError:
        print (response['message'])
        
```

```python
weather_list = list(zip(temperature_list,
humidity_list,
cloudiness_list,
wind_speed_list,
lat_list,
long_list,
city_list,
country_list))
# print(weather_list)
columns = ['Temperature','Humidity','Cloudiness','Wind Speed','Latitude','Longitude','City','Country']
city_weather_df = pd.DataFrame(weather_list, columns = columns)
# city_weather_df.head(10)
city_weather_df.info()
```

```python
#Save the CSV of the Weather data retrieved
city_weather_df.to_csv("city_weather.csv", header = True, index = False)
```
