Trends Observed:
1. As we get closer to Equator, the temperature rises.
2. There is no impact on the proximity of a city to its humidity and cloudiness
3. Cities farther from equator had higher wind speeds.

```python
import pandas as pd
import numpy as np
import requests, json
import matplotlib.pyplot as plt
from citipy import citipy
import openweathermapy.core as owm
```

```python
from config import api_key
```

```python
#Generate Random Lat and Long data to fetch random cities from citipy
random_lat = np.random.randint(-90, 91, 2000)
#print(len(random_lat))
random_long = np.random.randint(-180, 181, 2000)
# print(random_long)
```

```python
#Fetch the city details from Citipy library
random_cities = []
for gc in range(len(random_lat)):
    nearest_city = citipy.nearest_city(random_lat[gc], random_long[gc])
    if nearest_city not in random_cities:
        random_cities.append(nearest_city)
```

```python
#Initlialize lists to store the values returned by OWM API
temperature_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
lat_list = []
long_list = []
city_list = []
country_list = []
```

```python
#Fetch the weather data from OWM API
#set base url for OWM
base_url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid={api_key}&q="
# print (base_url)
for city_id in range(len(random_cities)):
    print(f"------Fetching weather info for city #{city_id}, {random_cities[city_id].city_name}------")
    complete_url = f"{base_url}{random_cities[city_id].city_name},{random_cities[city_id].country_code}"
    try:
        query = requests.get(complete_url)
        response = query.json()
        temperature_list.append(response['main']['temp_max'])
        humidity_list.append(response['main']['humidity'])
        cloudiness_list.append(response['clouds']['all'])
        wind_speed_list.append(response['wind']['speed'])
        lat_list.append(response['coord']['lat'])
        long_list.append(response['coord']['lon'])
        city_list.append(response['name'])
        country_list.append(response['sys']['country'])
    except KeyError:
        print (response['message'])
        
```

```python
#Zip the individual lists together and create a DF to store the values
weather_list = list(zip(temperature_list,
humidity_list,
cloudiness_list,
wind_speed_list,
lat_list,
long_list,
city_list,
country_list))
# print(weather_list)
columns = ['Temperature','Humidity','Cloudiness','Wind Speed','Latitude','Longitude','City','Country']
city_weather_df = pd.DataFrame(weather_list, columns = columns)
# city_weather_df.head(10)
city_weather_df.info()
```

```python
#Save the CSV of the Weather data retrieved
city_weather_df.to_csv("city_weather.csv", header = True, index = False)
```

```python
#Build Scatter Plots
#1. Temperature (F) vs. Latitude

plt.scatter(lat_list, temperature_list, marker="o", facecolors="blue", edgecolors="black", alpha=0.85)
plt.xlim(-80, 100)
plt.ylim(-50, 150)
plt.grid(True)
plt.title("City Latitude vs Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.savefig("LatitudeToMaxTemp.jpg")
plt.show()
```

```python
#Build Scatter Plots
#2. Humidity (%) vs. Latitude

plt.scatter(lat_list, humidity_list, marker="o", facecolors="blue", edgecolors="black", alpha=0.85)
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.grid(True)
plt.title("City Latitude vs Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.savefig("LatitudeToHumidity.jpg")
plt.show()
```

```python
#Build Scatter Plots
#3. Cloudiness (%) vs. Latitude

plt.scatter(lat_list, cloudiness_list, marker="o", facecolors="blue", edgecolors="black", alpha=0.85)
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.grid(True)
plt.title("City Latitude vs Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.savefig("LatitudeToCloudiness.jpg")
plt.show()
```

```python
#Build Scatter Plots
#4. Wind Speed (mph) vs. Latitude

plt.scatter(lat_list, wind_speed_list, marker="o", facecolors="blue", edgecolors="black", alpha=0.85)
plt.xlim(-80, 100)
plt.ylim(-5, 50)
plt.grid(True)
plt.title("City Latitude vs Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.savefig("LatitudeToWindspeed.jpg")
plt.show()
```
